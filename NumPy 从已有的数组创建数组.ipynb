# NumPy 从已有的数组创建数组

将列表转换为 ndarray:

```python
import numpy as np
x=[1,2,3]
a=np.asarray(x)
print(a)
```

将元组转换为 ndarray:

```python
x=(1,2,3)
a=np.asarray(x)
print(a)
```

将元组列表转换为 ndarray:

```python
x=[(1,2,3),(4,5)]
a=np.asarray(x,dtype=object)
print(a,a.shape)
```

设置了 dtype 参数：

```python
x=[1,2,3]
a=np.asarray(x,dtype=float)
print(a)
```

numpy.frombuffer


在Python3中，b前缀表示字节字符串。字节字符串是一种不同于普通字符串的数据类型，它存储的是原始字节而不是文本字符。在这个例子中，s被定义为一个字节字符串，因此它的值前面有一个b前缀。

当您打印ndarray a时，每个元素都是一个长度为1的字节字符串，因此每个元素前面也都有一个b前缀。

```python
s=b"hello world"
a=np.frombuffer(s,dtype="S1")
print(a)
```

numpy.fromiter


```python
# 使用range函数创建列表对象
list=range(5)
it=iter(list)

# 使用迭代器创建ndarray
x=np.fromiter(it,dtype=float)
print(x)
x=np.fromiter("hello world",dtype="U1")
print(x)
```
