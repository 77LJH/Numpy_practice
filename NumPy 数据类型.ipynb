# NumPy 数据类型

```python
import numpy as np

# 使用标量类型
dt=np.dtype(np.int32)
dt
```

```python
# int8, int16, int32, int64 四种数据类型可以使用字符串 'i1', 'i2','i4','i8' 代替
dt=np.dtype("i4")
dt
```

```python
# 字节顺序标注
dt=np.dtype("<i4")
dt
```

```python
# 首先创建结构化数据类型
dt=np.dtype([("age",np.int8)])
dt
```

```python
# 将数据类型应用于ndarray对象
dt=np.dtype([("age",np.int8)])
a=np.array([(10,),(20,),(30,)],dtype=dt)
a
```

```python
# 类型字段名可以用于存取实际的age列
dt=np.dtype([("age",np.int8)])
a=np.array([(10,),(20,),(30,)],dtype=dt)
a["age"]
```

下面的示例定义一个结构化数据类型 student，包含字符串字段 name，整数字段 age，及浮点字段 marks，并将这个 dtype 应用到 ndarray 对象。

```python
student=np.dtype([("name","S20"),("age","i1"),("marks","f4")])
student
```

这段代码定义了一个名为 student 的结构化数据类型，其中包含三个字段：name（字符串，最大长度为 20 个字符），age（8 位整数）和 marks（32 位浮点数）。然后使用这个数据类型创建了一个结构化 numpy 数组 a，并将其填充了两个元素，每个元素都是一个元组，包含三个值。最后打印数组 a 的结果。

下面是一个表格，它可以帮助您更好地理解这段代码的结果：


```python
student=np.dtype([("name","S20"),("age","i1"),("marks","f4")])
a=np.array([("lihua",21,89),("xiaom",22,56)],dtype=student)
print(a)
print(a["name"])
```

```python
Pig=np.dtype([("kind","S20"),("age","i1"),("weight","f4"),("habit","S50")])
pig=np.array([("zhuzhu",2,60,"run and sleep"),("zz",3,89,"watch TV and sing")],dtype=Pig)
pig
```
