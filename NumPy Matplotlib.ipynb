# NumPy Matplotlib

Matplotlib 是 Python 的绘图库。 它可与 NumPy 一起使用，提供了一种有效的 MatLab 开源替代方案。 它也可以和图形工具包一起使用，如 PyQt 和 wxPython。

```python
import numpy as np 
from matplotlib import pyplot as plt 
 
x = np.arange(1,11) 
y =  2  * x +  5 
plt.title("Matplotlib demo") 
plt.xlabel("x axis caption") 
plt.ylabel("y axis caption") 
plt.plot(x,y) 
plt.show()
```

```python
# 计算正弦曲线上点的 x 和 y 坐标
x = np.arange(0,  3  * np.pi,  0.1) 
y = np.sin(x)
plt.title("sine wave form")  
# 使用 matplotlib 来绘制点
plt.plot(x, y) 
plt.show()
```

subplot()
import matplotlib.pyplot as plt

fig, axs = plt.subplots(nrows, ncols, sharex=False, sharey=False, figsize=(width, height), gridspec_kw=None)

其中，fig是整个图形的Figure对象，axs是一个形如nrows×ncols的Axes对象数组，每个Axes对象代表一个子图。

subplot() 函数允许你在同一图中绘制不同的东西。

以下实例绘制正弦和余弦值:



```python
# 计算正弦和余弦曲线上的点的 x 和 y 坐标 
x = np.arange(0,  3  * np.pi,  0.1) 
y_sin = np.sin(x) 
y_cos = np.cos(x)  

# 建立 subplot 网格，长为 2，宽为 1  

# 激活第一个 subplot
plt.subplot(2,  1,  1)  
# 绘制第一个图像 
plt.plot(x, y_sin) 
plt.title('Sine')  
# 将第二个 subplot 激活，并绘制第二个图像
plt.subplot(2,  1,  2) 
plt.plot(x, y_cos) 
plt.title('Cosine')  
# 展示图像
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 生成数据
x = np.linspace(0, 2*np.pi, 100) # 等差数列
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.exp(-x)

# 创建子图
fig, axs = plt.subplots(nrows=2, ncols=2, sharex=False, sharey=False, figsize=(8, 6))

# 绘制第1个子图
axs[0, 0].plot(x, y1)
axs[0, 0].set_title('sin(x)')

# 绘制第2个子图
axs[0, 1].plot(x, y2)
axs[0, 1].set_title('cos(x)')

# 绘制第3个子图
axs[1, 0].plot(x, y3)
axs[1, 0].set_title('tan(x)')

# 绘制第4个子图
axs[1, 1].plot(x, y4)
axs[1, 1].set_title('exp(-x)')

# 设置整个图形的标题
fig.suptitle('Four functions')

# 显示图形
plt.show()
```

bar()
pyplot 子模块提供 bar() 函数来生成条形图。

以下实例生成两组 x 和 y 数组的条形图。

在 Matplotlib 的 bar 函数中，align 参数控制每个条形图的位置。

如果设置为 'center'，则每个条形图将放置在其指定位置的中心。如果设置为 'edge'，则每个条形图将放置在其指定位置的边缘。

例如，当 align='center' 时，如果一个条形图的宽度为 0.5，那么它将从给定位置的 0.25 到 0.75。

```python
from matplotlib import pyplot as plt 
x =  [5,8,10] 
y =  [12,16,6] 
x2 =  [6,9,11] 
y2 =  [6,15,7] 
plt.bar(x, y, align =  'center') 
plt.bar(x2, y2, color =  'g', align =  'center') 
plt.title('Bar graph') 
plt.ylabel('Y axis') 
plt.xlabel('X axis') 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 设置数据
x = np.array(['A', 'B', 'C', 'D'])
y = np.array([20, 35, 30, 25])

# 创建柱状图
plt.bar(x, y)

# 添加标签和标题
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Vertical Bar Chart')

# 展示图像
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 创建数据
x = np.arange(1, 6)
y = [3, 2, 5, 4, 1]

# 绘制两个 bar 图，一个使用 'edge'，一个使用 'center' 对齐方式
fig, axs = plt.subplots(1, 2, figsize=(8, 4))

axs[0].bar(x, y, width=0.4, align='edge')
axs[0].set_title("align='edge'")
axs[1].bar(x, y, width=0.4, align='center')
axs[1].set_title("align='center'")

plt.show()
```

numpy.histogram()
numpy.histogram() 函数是数据的频率分布的图形表示。 水平尺寸相等的矩形对应于类间隔，称为 bin，变量 height 对应于频率。

numpy.histogram()函数将输入数组和 bin 作为两个参数。 bin 数组中的连续元素用作每个 bin 的边界。

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27])
np.histogram(a,bins =  [0,20,40,60,80,100]) 
hist,bins = np.histogram(a,bins =  [0,20,40,60,80,100])  
print (hist) 
print (bins)
```

plt()
Matplotlib 可以将直方图的数字表示转换为图形。 pyplot 子模块的 plt() 函数将包含数据和 bin 数组的数组作为参数，并转换为直方图。

```python
a = np.array([22,87,5,43,56,73,55,54,11,20,51,5,79,31,27]) 
plt.hist(a, bins =  [0,20,40,60,80,100]) 
plt.title("histogram") 
plt.show()
```
