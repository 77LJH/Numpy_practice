# NumPy 数组属性

ndarray.ndim 用于返回数组的维数，等于秩。

```python
import numpy as np

a=np.arange(24)
print(a,    a.ndim) # a 现只有一个维度

#现在调整其大小
b=a.reshape(2,4,3) # 现在b有三个维度
print(b,    b.ndim)
```

ndarray.shape 表示数组的维度，返回一个元组，这个元组的长度就是维度的数目，即 ndim 属性(秩)。比如，一个二维数组，其维度表示"行数"和"列数"。

ndarray.shape 也可以用于调整数组大小。

```python
a=np.array([[1,2,3],[4,5,6]])
a.shape
```

```python
a=np.array([[1,2,3],[4,5,6]])
a.shape=(3,2)
a
```

ndarray.itemsize 以字节的形式返回数组中每一个元素的大小。

例如，一个元素类型为 float64 的数组 itemsize 属性值为 8(float64 占用 64 个 bits，每个字节长度为 8，所以 64/8，占用 8 个字节），又如，一个元素类型为 complex32 的数组 item 属性为 4（32/8）。

```python
# 数组的dtype为int8（"i1"）
x=np.array([1,2,3,4,5],dtype=np.int8)
print(x.itemsize)

# 数组的dtype现在为 float64(8个字节)
y=np.array([1,2,3,4,5],dtype=np.float64)
print(y.itemsize)

z=np.array([[1,2,3],[4,5,6]],dtype=np.int8)
print(z.itemsize)
```

```python
x=np.array([1,2,3,4,5])
print(x.flags)
```

ndarray.reshape 通常返回的是非拷贝副本，即改变返回后数组的元素，原数组对应元素的值也会改变。
ndarray.reshape 函数用于改变数组的形状，它返回一个新的数组，其形状与指定的形状相同，但数据仍然与原始数组共享。这意味着返回的数组并不是原始数组的拷贝副本，而是一个视图。

由于返回的数组与原始数组共享数据，因此当您修改返回的数组中的元素时，原始数组中对应的元素也会发生变化。同样，如果您修改原始数组中的元素，返回的数组中对应的元素也会发生变化。

```python
a=np.array([[1,2,3],[4,5,6]])
print(a)
b=a.reshape((6,))
print(b)
b=a.reshape(6)
print(b)

b[0]=100
print(b)
print(a)
```
